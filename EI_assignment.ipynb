```python
import pandas as pd

df = pd.read_excel('/content/drive/MyDrive/UniBradStudy/cx_dataset.xlsx')
```

```python
df.head()
```

```python
School_fees = len(df[df['Reason for cx'].str.contains('Fees')])
```

```python
Remitance = len(df[df['Reason for cx'].str.contains('Remitance')])
```

```python
Upkeep = len(df[df['Reason for cx'].str.contains('Upkeep')])
```

```python
import matplotlib.pyplot as plt
counts = [School_fees,Remitance,Upkeep]
labels = ['School fess',' Remitance', 'Upkeep']

plt.bar(labels, counts)
plt.xlabel('Use of currencey')
plt.ylabel('Users')
plt.title('Why people need currency exchange')
plt.show()
```
